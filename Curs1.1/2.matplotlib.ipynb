# functia scatter - pune punctele imprastiate

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]
plt.scatter(lista_1, lista_2);
plt.show()
```

```python
lista_1 = [1]
lista_2 = [4]
plt.scatter(lista_1, lista_2);
plt.show()
```

# matplot poate lua si range-uri (ArrayLike)

```python
range_1 = range(4, 8)       # 4,5,6,7,8
range_2 = range (6, 10)     # 6,7,8,9,10

plt.scatter(range_1, range_2);
plt.show()
```

```python
range_1 = [3] * 4
range_2 = range (4)     

plt.scatter(range_1, range_2);
plt.show()
```

```python
x = [10, 10]
y = [8, 9]

plt.scatter(x, y)
```

# plt.title -> Titlu

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu")
plt.scatter(x, y)
```

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu")
plt.xlabel("Valoarea x")
plt.scatter(x, y);
```

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu")
plt.xlabel("Valoarea x")
plt.ylabel("Valoarea y")
plt.scatter(x, y);
```

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu")
plt.xlabel("Valoarea x")
plt.ylabel("Valoarea y")
plt.scatter(x, y);

plt.savefig("Figura_mea.png")
```
